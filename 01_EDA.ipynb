# 📊 Notebook 1 — Exploratory Data Analysis
**Customer Churn Predictor**

In this notebook we explore the Telco Churn dataset to understand:
- Distribution of churned vs non-churned customers
- Which features correlate most with churn
- Patterns in contract type, tenure, charges

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid', palette='muted')
%matplotlib inline
```

```python
# Load data
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 1. Basic Info & Missing Values

```python
df.info()
print('\nMissing values:')
print(df.isnull().sum())

# Fix TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f'\nRows with NaN TotalCharges: {df["TotalCharges"].isnull().sum()}')
```

## 2. Churn Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Count plot
df['Churn'].value_counts().plot(kind='bar', ax=axes[0], color=['steelblue', 'tomato'], rot=0)
axes[0].set_title('Churn Count')
axes[0].set_xlabel('Churn')
axes[0].set_ylabel('Count')

# Pie chart
df['Churn'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%', colors=['steelblue', 'tomato'])
axes[1].set_title('Churn Rate')
axes[1].set_ylabel('')

plt.tight_layout()
plt.savefig('../reports/churn_distribution.png', dpi=150)
plt.show()
```

## 3. Churn by Contract Type

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Contract', hue='Churn', palette={'No': 'steelblue', 'Yes': 'tomato'})
plt.title('Churn by Contract Type')
plt.savefig('../reports/churn_by_contract.png', dpi=150)
plt.show()
```

## 4. Tenure vs Monthly Charges (coloured by Churn)

```python
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='tenure', y='MonthlyCharges', hue='Churn',
                palette={'No': 'steelblue', 'Yes': 'tomato'}, alpha=0.6)
plt.title('Tenure vs Monthly Charges')
plt.savefig('../reports/tenure_vs_charges.png', dpi=150)
plt.show()
```

## 5. Correlation Heatmap

```python
# Encode for correlation
df_enc = df.copy()
df_enc['TotalCharges'].fillna(df_enc['TotalCharges'].median(), inplace=True)
df_enc['Churn'] = df_enc['Churn'].map({'Yes': 1, 'No': 0})
for col in df_enc.select_dtypes(include='object').columns:
    df_enc[col] = df_enc[col].astype('category').cat.codes

plt.figure(figsize=(14, 10))
sns.heatmap(df_enc.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../reports/correlation_heatmap.png', dpi=150)
plt.show()
```

## ✅ Key Findings

Write your observations here after running the notebook:

- **Churn rate:** ~26% of customers have churned
- **Contract type:** Month-to-month customers churn most
- **Tenure:** Short-tenure customers are more likely to churn
- **Monthly charges:** Higher monthly charges correlate with higher churn
